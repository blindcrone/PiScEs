# PiScEs: A simple pixel scale estimator

```python
import os
import pisces
from PIL import Image
from time import time
import numpy as np
from glob import glob
```

```python
estimate = pisces.PiScEs(4, 0x64, 4)
samples = glob('samples/*')
example = np.random.choice(samples, 1)[0]
base = Image.open(example)
base
```

```python
img = base.crop((0,0,1024,1008))
t0 = time()
scale = estimate(img)
time() - t0
```

```python
scale
```

```python
resized = img.resize((img.size[0] // scale, img.size[1] // scale,))
resized
```
